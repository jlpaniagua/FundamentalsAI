##**Primer grafico**

```
import matplotlib.pyplot as plt
plt.plot([1, 2, 4, 9, 5, 3])
plt.show()
```

* **plot** es la funcion encargada de contruir el grafico
*  **show** es la funcion que permite mostrar el grafico

* la funcion **plot** recibe los datos en forma de vector




```
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.show()
```

El primer vector o parametro de la funcion plot hace referencia al eje horizontal del grafico. El segundo especifica el eje vertical 

```
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.axis([-4, 6, 0, 7])
plt.show()
```

la funcion **axis** permite configurar el tamaño de los ejes del grafico

```
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

A continuacion se agregaran un titulo, los nombres de los y una grilla o cuadricula.

```
plt.plot(x, y)
plt.title("Square function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

# **Estilo y Corlor**

por defecto, matlplotlib dibuja una linea entre puntos conseecutivos

```
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0])
plt.axis([-10, 110, -10, 140])
plt.show()
```

existe un tercer argumento o parametro, el cual indica el color y el estilo de la linea del grafico

```
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

es posible graficar lineas con diferente estilo dentro de un mismo grafico.

```
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-", [0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

una forma mas ordenada se graficar varios datos en un mismo grafico se muestra a continuacion

```
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-")
plt.plot([0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

en lugar de dibujar lineas, es posible graficar los datos utilizando puntos con diferente forma y color

```
x = np.linspace(-1.4, 1.4, 30)
plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
plt.show()
```

existes un gran numero de atributos que se puede modficar para logar una apariencia deseada en el grafico

```
x = np.linspace(-1.4, 1.4, 30)
line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
line1.set_linewidth(3.0)
line1.set_dash_capstyle("round")
line3.set_alpha(0.2)
plt.show()
```

## **Dibujar varios graficos en una misma figura (Subplots)**

* Para crear una figura con multiples graficos se utiliza la funcion **subplot**, la cual permite organizar dichos graficos en una cuadricula (grid).

* La funcion subplot recibe como parametros la cantidad de filas y columnas de la figura y el indice de la grafica que se desea dibujar en dicha posicion.

* Las posiciones se indican de izquierda a derecha y de arriba a abajo.

```
x = np.linspace(-1.4, 1.4, 30)
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
plt.plot(x, x**2)
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, 3rd subplot = bottow left
plt.plot(x, x**3)
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, 4th subplot = bottom right
plt.plot(x, x**4)
plt.show()
```

Algunos graficos necesitan extenderse por mas de una posicion en la cuadricula 

```
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
plt.plot(x, x**2)
plt.subplot(2, 1, 2)  # 2 rows, *1* column, 2nd subplot = bottom
plt.plot(x, x**3)
plt.show()
```

para posicionar graficos dentro de una misma figura de una forma mas compleja, se puede utilizar la funcion **subplot2gird**. En este caso se deben especificar el numero de filas y columnas en la cuadricula, la position de la grafica dentro de la cuadricula, y opcionalmente cuantas filas y columnas ocupara dicha grafica.





```
plt.subplot2grid((3,3), (0, 0), rowspan=2, colspan=2)
plt.plot(x, x**2)
plt.subplot2grid((3,3), (0, 2))
plt.plot(x, x**3)
plt.subplot2grid((3,3), (1, 2), rowspan=2)
plt.plot(x, x**4)
plt.subplot2grid((3,3), (2, 0), colspan=2)
plt.plot(x, x**5)
plt.show()
```

## **Multiples Figuras**

la funcion **figure** permite crear diferentes figuras que pueden contener diferentes graficos. Es es util cuando se quieren guardar graficos en diferentes figuras.

```
x = np.linspace(-1.4, 1.4, 30)

plt.figure(1)
plt.subplot(211)
plt.plot(x, x**2)
plt.title("Square and Cube")
plt.subplot(212)
plt.plot(x, x**3)

plt.figure(2, figsize=(10, 5))
plt.subplot(121)
plt.plot(x, x**4)
plt.title("y = x**4")
plt.subplot(122)
plt.plot(x, x**5)
plt.title("y = x**5")

plt.figure(1)      # back to figure 1, current subplot is 212 (bottom)
plt.plot(x, -x**3, "r:")

plt.show()
```

Existe una forma mas interactiva para crear graficos, si necesidad de utiliza siempre la funcion plot. Pero recordar que una de las filosofias de Python es que explicito es mejor que implicito

```
x = np.linspace(-2, 2, 200)
fig1, (ax_top, ax_bottom) = plt.subplots(2, 1, sharex=True)
fig1.set_size_inches(10,5)
line1, line2 = ax_top.plot(x, np.sin(3*x**2), "r-", x, np.cos(5*x**2), "b-")
line3, = ax_bottom.plot(x, np.sin(3*x), "r-")
ax_top.grid(True)

fig2, ax = plt.subplots(1, 1)
ax.plot(x, x**2)
plt.show()
```
