2. La solución de una ecuación diferencial esta dada por la siguiente ecuación 
y(t)=C-exp(-r*t) 
Grafique la respuesta del sistema con C=5 y r=2

```
t=np.linspace(0, 5, 100)
y=5-np.exp(-2*t)
plt.pyplot.plot(t, y)
plt.pyplot.show()
```

3. La solución de una ecuación diferencial esta dada por la siguiente ecuación 
y(t)=C-exp(-r*t) 
Grafique la respuesta del sistema con C=5 y r con valores de 1 a 5

```
t=np.linspace(0, 5, 100)
yt= np.zeros((5,100))
for i in range(1,5):
  yt[i,:]=5-np.exp(-i*t)

for i in range(1,5):
  plt.pyplot.plot(t,yt[i,:])
  plt.pyplot.show()


```

4. Simular la siguiente ecuación W(k+1)=W(k)+p DW(k) que implementa el proceso de entrenamiento de una red neuronal artificial durante 100 iteraciones. Suponga p=0.9 y Dw(k)= valor randómico


```
W= np.zeros((100,1))
for k in range(0,99):
  W[k+1]=W[k]+0.9*np.random.randn()

plt.pyplot.plot(W)
plt.pyplot.show()
```

5. Realizar algún ejercicio con manipulación de imágenes cargándolas como matrices en numpy

```
import matplotlib.image as mpimg

img = mpimg.imread('index.jpeg')
print(img.shape, img.dtype)
```

```
plt.pyplot.imshow(img)
plt.pyplot.show()
```

```
plt.pyplot.imshow(img)
plt.pyplot.axis('off')
plt.pyplot.show()
```
